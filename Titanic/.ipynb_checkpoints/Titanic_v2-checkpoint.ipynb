```python
import pandas as pd
import numpy as np

df_train = pd.read_csv('./train.csv')
df_test = pd.read_csv('./test.csv')

print(df_test.shape)
print(df_test.columns)
print(df_test.dtypes)

print("\n\n")
print(df_train.shape)
print(df_train.columns)
print(df_train.dtypes)
print(df_train.head())
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train['Cabin'].replace(np.nan,'None',inplace=True)
df_test['Cabin'].replace(np.nan,'None',inplace=True)
```

```python
import re
df_train['Cab_level'] = df_train['Cabin'].apply(lambda x: (re.search(r'[a-zA-Z]{1,2}',x)).group(0))
df_train.head()
```

```python
df_test['Cab_level'] = df_test['Cabin'].apply(lambda x: (re.search(r'[a-zA-Z]{1,2}',x)).group(0))
df_test.head()
```

```python
from sklearn.impute import SimpleImputer
df_train[['Age']] = SimpleImputer(np.nan,strategy = 'mean').fit_transform(df_train[['Age']])
df_train[['Embarked']] = SimpleImputer(np.nan,strategy = 'most_frequent').fit_transform(df_train[['Embarked']])
df_train[['Cab_level']] = SimpleImputer('No',strategy = 'most_frequent').fit_transform(df_train[['Cab_level']])

df_test[['Age']] = SimpleImputer(np.nan,strategy = 'mean').fit_transform(df_test[['Age']])
df_test[['Fare']] = SimpleImputer(np.nan,strategy = 'mean').fit_transform(df_test[['Fare']])
df_test[['Cab_level']] = SimpleImputer('No',strategy = 'most_frequent').fit_transform(df_test[['Cab_level']])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_test['Sex_enc'] = encoder.fit_transform(df_test['Sex'])

#df_test['Embark_enc']= encoder.fit_transform(df_test['Embarked'])

#df_test['Cab_enc']= encoder.fit_transform(df_test['Cab_level'])


df_train['Sex_enc'] = encoder.fit_transform(df_train['Sex'])

#df_train['Embark_enc']= encoder.fit_transform(df_train['Embarked'])

#df_train['Cab_enc']= encoder.fit_transform(df_train['Cab_level'])
```

```python
features_select = ['Pclass','Sex_enc','Age','SibSp','Parch','Fare']
X = df_train.drop('Survived',axis=1)
Y = df_train['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42, solver='lbfgs',max_iter=1000).fit(X_train[features_select],Y_train)
Y_pred = clf.predict(X_test[features_select])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer()
df_test_Embark = pd.DataFrame(encoder.fit_transform(df_test['Embarked']),columns = encoder.classes_,dtype=float)

df_test_Cab = pd.DataFrame(encoder.fit_transform(df_test['Cab_level']),columns = encoder.classes_,dtype=float)

df_test = pd.concat([df_test,df_test_Embark],axis=1,sort=False)
df_test = pd.concat([df_test,df_test_Cab],axis=1, sort=False)

df_train_Embark = pd.DataFrame(encoder.fit_transform(df_train['Embarked']),columns = encoder.classes_,dtype=float)

df_train_Cab = pd.DataFrame(encoder.fit_transform(df_train['Cab_level']),columns = encoder.classes_,dtype=float)

df_train = pd.concat([df_train,df_train_Embark],axis=1,sort=False)
df_train = pd.concat([df_train,df_train_Cab],axis=1, sort=False)
```

```python
df_train.head()
```

```python
features_select = ['Pclass','Sex_enc','Age','SibSp','Parch','Fare','C','Q','S','A','B','C','D','E','F','G','T']
X = df_train.drop('Survived',axis=1)
Y = df_train['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42, solver='lbfgs',max_iter=1000).fit(X_train[features_select],Y_train)
Y_pred = clf.predict(X_test[features_select])
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid2 = [{'n_estimators':[10,20,50,100], 'criterion':['gini','entropy'], 
               'oob_score':[True, False]}]
clf2 = RandomForestClassifier(n_jobs=-1, random_state=42)
grid_search2 = GridSearchCV(clf2, param_grid2, cv=5, verbose=3)
#verbose option for display more logs on the screen, slowdown the algorithm
grid_search2.fit(X_train[features_select],Y_train)
```

```python
grid_search2.best_params_
```

```python
grid_search2.best_score_
```

```python
Y_pred2 = grid_search2.predict(X_test[features_select])
accuracy_score(Y_test, Y_pred2)
```
